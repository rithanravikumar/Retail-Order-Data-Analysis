```python
import pandas as pd
```

```python
df = pd.read_csv('/content/orders.csv')
```

```python
df
```

```python
print(list(df.columns))
```

```python
df.columns=['order_id', 'order_date', 'ship_mode', 'segment', 'country', 'city', 'state', 'postal_code','category','sub_category', 'product_id', 'region', 'cost_price', 'list_price', 'quantity', 'discount_percent']
```

```python
df.head()
```

```python
df['discount_price'] = df['list_price'] * (df['discount_percent'] / 100)
df['sale_price'] = df['list_price'] - df['discount_price']
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df["year"]=df["order_date"].dt.year
df["month"]=df["order_date"].dt.month
```

```python
print(list(df.columns))
```

```python
df.columns=['order_id', 'order_date', 'ship_mode', 'segment', 'country', 'city', 'state', 'postal_code','region','category','sub_category', 'product_id', 'cost_price', 'list_price', 'quantity', 'discount_percent','discount_price','sale_price','profit','year','month']
```
